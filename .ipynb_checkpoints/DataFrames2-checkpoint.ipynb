```python
from time import time
from pandas import DataFrame as DF
import numpy as np

data = DF()

np.random.seed(int(time()))

data["Nombre"] = ["Pedro", "Yvonne", "Ernesto", "Nadia", "Delfina", "Rodrigo"]
data["Estatura (m)"] = [round(np.random.normal(1.72, 0.08), 2) for persona in data["Nombre"]]
data["Peso (kg)"] = np.random.uniform(68, 110, len(data["Nombre"]))
data["Peso (kg)"] = [round(peso, 1) for peso in data["Peso (kg)"]]
data["IMC"] = round(data["Peso (kg)"] / data["Estatura (m)"] ** 2, 1)

data
```

```python
data_obesidad = data[data["IMC"] > 30].reset_index(drop=True)

data_obesidad
```
